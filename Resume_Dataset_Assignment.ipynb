```python
from sklearn.model_selection import train_test_split

```

*This line imports the train_test_split function from the sklearn.model_selection module. This function is used to split data into different subsets.*

```python
data = data[['Resume_str', 'Category']].rename(columns={'Resume_str': 'resume_str', 'Category': 'category'})
```

This line selects two columns, 'Resume_str' and 'Category', from the original dataset (data) and assigns them to a new DataFrame named data.

Then, it uses the .rename() method to rename the columns to 'resume_str' and 'category', respectively. This improves code readability by using lowercase and consistent naming conventions. *italicized text*

```python
# Train-Temporary Split: Split the data into train and temporary sets (80-20)
train_data, temp_data = train_test_split(data, test_size=0.2, random_state=42)
```

This line splits the data DataFrame into two new DataFrames: train_data and temp_data.
The train_test_split function is used, and the following arguments are provided:

* data: The original DataFrame to be split.

* test_size=0.2: This specifies that 20% (0.2) of the data should be allocated to the temporary set (temp_data) and the remaining 80% to the training set (train_data).

* random_state=42: This sets a seed for the random number generator used to shuffle the data before splitting. This ensures reproducibility (getting the same split each time you run the code). *italicized text*

```python
# Test-Validation Split: Split the temporary set into test and validation sets (50-50)
test_data, validation_data = train_test_split(temp_data, test_size=0.5, random_state=42)
```

This line further splits the temp_data DataFrame into two new DataFrames: test_data and validation_data.

The same train_test_split function is used, but this time:
* temp_data: The temporary DataFrame is used as the input data to be split.

* test_size=0.5: This specifies that 50% of the data in temp_data is allocated to the testing set (test_data), and the remaining 50% to the validation set (validation_data). *italicized text*

* The from_pandas() method converts the provided DataFrame into a Hugging Face Dataset object and assigns it to a variable (e.g., train_dataset). *italicized text*

# 4 load dataset from Huggingface

```python
from datasets import load_dataset, load_from_disk, load_metric
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
```

```python
# https://huggingface.co/datasets/pallavi176/resume_dataset
# Step 1: Load dataset from hub
data = load_dataset("Harsshadpawar06/resume_dataset1")
data
```

```python
data['train']
```

```python
#Step 2: Load a tokenizer and model from the Hugging Face Model Hub
model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
# Create label mappings
unique_labels = set(data['train']['category'] + data['test']['category'] + data['validation']['category'])
label2id = {label: id for id, label in enumerate(unique_labels)}
id2label = {id: label for label, id in label2id.items()}
```

```python
# Step 3: Tokenize the dataset
def tokenize_and_encode(ds):
    tokenized = tokenizer(ds['resume_str'], padding='max_length', truncation=True, max_length=512)
    tokenized['labels'] = [label2id[label] for label in ds['category']]
    return tokenized

tokenized_datasets = data.map(tokenize_and_encode, batched=True)
```

```python
tokenized_datasets
```

```python
tokenized_datasets['train']['resume_str'][0]
```

```python
tokenized_datasets['train']['category'][0]
```

```python
# Load the model with the custom label configuration
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint, num_labels=len(unique_labels), id2label=id2label, label2id=label2id)
```

```python
# Step 4: Fine-tune the model
training_args = TrainingArguments(
    output_dir="./results",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=1,
    weight_decay=0.01,
    evaluation_strategy="epoch",
)
```

```python
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='macro')
    acc = accuracy_score(labels, preds)
    return {'accuracy': acc, 'f1': f1, 'precision': precision, 'recall': recall}
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
# Save the model and tokenizer
model.save_pretrained('./fine_tuned_model')
tokenizer.save_pretrained('./fine_tuned_model_tokens')
```

```python
# Evaluate the model
eval_results = trainer.evaluate()
eval_results
```
